```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
#Filter by countries
#df_canada.OdName
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_canada_2.columns = list(map(str, df_canada_2.columns)) # to prevent ambiguity between index numbers and column names
```

```python
df_canada_2.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
df_canada_2
```

```python
# Get the total population for each year
df_total = pd.DataFrame(df_canada_2[years].sum(axis=0))
df_total.index = map(int, df_total.index)
df_total.reset_index(inplace = True)

# rename columns to year and total
df_total.columns = ['year', 'total']

#show example 
df_total.head()
```

```python
# Draw the line of best fit
x = df_total['year']      # year on x-axis
y = df_total['total']     # total on y-axis
fit = np.polyfit(x, y, deg=1)
```

```python
df_total.plot(kind='scatter', x='year', y='total')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# Linear Regression Equation
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```
