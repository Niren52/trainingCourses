```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
#Filter by countries
#df_canada.OdName
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_canada_2.columns = list(map(str, df_canada_2.columns)) # to prevent ambiguity between index numbers and column names
```

```python
df_canada_2.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
df_canada_2
```

```python
df_canada_t = df_canada_2[years].transpose() # transposed dataframe containing just yearly data

# Make the years datatype into integer
df_canada_t.index = map(int, df_canada_t.index)

# let's label the index. This will automatically be the column name when we reset the index
df_canada_t.index.name = 'Year'

# reset index to bring the Year in as a column
df_canada_t.reset_index(inplace=True)

# view the changes
df_canada_t.head()
```

### Scale all the data with the following information to get data between 0 and 1 using the following equation


```python
# normalise data for Argentina and Brazil
norm_brazil = (df_canada_t['Brazil'] - df_canada_t['Brazil'].min()) / (df_canada_t['Brazil'].max() - df_canada_t['Brazil'].min())
norm_argentina = (df_canada_t['Argentina'] - df_canada_t['Argentina'].min()) / (df_canada_t['Argentina'].max() - df_canada_t['Argentina'].min())
```

```python
# Plot the data

# Brazil
ax0 = df_canada_t.plot(kind='scatter',x='Year', y='Brazil', figsize=(14, 8), alpha=0.5, color='green', s=norm_brazil * 2000 + 10, xlim=(1975, 2015))

# Argentina
ax1 = df_canada_t.plot(kind='scatter', x='Year', y='Argentina', alpha=0.5, color="blue", s=norm_argentina * 2000 + 10, ax = ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
