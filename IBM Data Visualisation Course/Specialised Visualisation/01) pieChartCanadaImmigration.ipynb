```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
#Filter by countries
#df_canada.OdName
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_canada_2.columns = list(map(str, df_canada_2.columns)) # to prevent ambiguity between index numbers and column names
```

```python
df_canada_2.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
df_canada_2
```

```python
# Group countries by continent and take all the totals

df_continents = df_canada_2.groupby('AreaName', axis=0).sum()
df_continents # Get data
```

```python
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         
                            pctdistance=1.12,    
                            explode=explode_list # enlarge the segments for lowest segments
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_continents.index, loc='lower right') 

plt.show()
```
