```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
#Filter by countries
#df_canada.OdName
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_canada_2.columns = list(map(str, df_canada_2.columns)) # to prevent ambiguity between index numbers and column names
```

```python
df_canada_2.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
df_canada_2
```

```python
# Get Japan data as a dataframe
df_japan = df_canada_2.loc[['Japan'], years].transpose()
df_japan.head()
```

```python
df_japan.plot(kind='box', figsize=(8, 6))

plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
# Get Summary data Statistics

df_japan.describe()
```

```python
df_china_india= df_canada_2.loc[['China', 'India'], years].transpose()
df_china_india.head()
```

```python
#Compare Statsistics
df_china_india.describe()
```

```python
# Plots for China and India
df_china_india.plot(kind='box')

plt.title('Box plots of Immigrants from China and India (1980 - 2013)')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# Show Plots of two different graphs together
fig = plt.figure() 

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

# Subplot 1: Box plot
df_china_india.plot(kind='box', vert=False, ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
df_china_india.plot(kind='line', ax=ax1) # add to subplot 2
ax1.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
```
