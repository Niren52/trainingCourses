```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
#Filter by countries
#df_canada.OdName
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_canada_2.columns = list(map(str, df_canada_2.columns)) # to prevent ambiguity between index numbers and column names
```

```python
df_canada_2.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
df_canada_2
```

```python
# Get Iceland Data
df_iceland = df_canada_2.loc['Iceland', years]
df_iceland.head()
```

```python
# Plot graph to show barchart
df_iceland.plot(kind='bar')

plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')



plt.show()
```
