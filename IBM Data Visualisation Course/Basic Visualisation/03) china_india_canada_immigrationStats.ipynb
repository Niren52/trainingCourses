```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
#Filter by countries
df_canada.OdName
```

```python
df_canada_2.columns = list(map(str, df_canada_2.columns)) # to prevent ambiguity between index numbers and column names
```

```python
df_canada_2
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_india_china = df_canada_2.loc[['India', 'China'], years] # Get the data for india and China
```

```python
 df_india_china = df_india_china.transpose() # done to make graphing compatible with matplotlib
```

```python
df_india_china.head()
```

```python
#Plot the graph

df_india_china.index = df_india_china.index.map(int)
df_india_china.plot(kind='line')


plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```
