```python
%matplotlib notebook
#Import Needed Modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Read the file
df_canada = pd.read_excel('Canada.xlsx',
                         sheet_name="Canada by Citizenship",
                         skiprows=range(20),
                         skipfooter=2)
```

```python
# Get summary of dataframe
df_canada.head()
```

```python
df_canada["Total"] = df_canada[range(1980,2014)].sum(axis=1)
```

```python
df_canada_2 = df_canada.set_index('OdName')
```

```python
df_canada_2.head()
```

```python
# Years we want to plot data for
years = list((range(1980,2014)))
```

```python
#Plot the graph
df_canada_2.loc['Haiti', years].plot(kind = 'line')
plt.title('Immigration of Haitians to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
