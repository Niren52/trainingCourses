```python
%matplotlib notebook
# Import modules for plotting, wordcloud creation, and mask creation
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
# Import 'Alice's Adventures In Wonderland' to notebook
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
alice_novel
```

```python
# Create a mask of the alice Black and White Image
alice_mask = np.array(Image.open('alice_mask.png'))
```

```python
# Remove commonly used stop words such as; 'It', 'The' 'They', so we can get more meaning from data
stopwords = set(STOPWORDS)
```

# Refined Word Cloud without 'Said'

```python
# Remove stop words 'Said' because it doesn't add much insight, and then regenerate word cloud

stopwords.add('said') 

# Create the wordCloud and set parameters
alice_wc = WordCloud(
    background_color='white',
    mask=alice_mask,
    max_words=2000, # Maximum of 2000 word to prevent Word cloud from getting less insightful
    stopwords=stopwords
)

# Create word cloud
alice_wc.generate(alice_novel)

# Plot the word cloud
fig = plt.figure()
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
plt.savefig('aliceWordCloud2.png')
```
