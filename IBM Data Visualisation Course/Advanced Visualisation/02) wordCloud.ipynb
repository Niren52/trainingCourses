```python
%matplotlib notebook
# Import modules for plotting and wordcloud creation
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# Import 'Alice's Adventures In Wonderland' to notebook
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
alice_novel
```

```python
# Remove commonly used stop words such as; 'It', 'The' 'They', so we can get more meaning from data
stopwords = set(STOPWORDS)
```

```python
# Create the wordCloud and set parameters
alice_wc = WordCloud(
    background_color='white',
    max_words=2000, # Maximum of 2000 word to prevent Word cloud from getting less insightful
    stopwords=stopwords
)

# Create word cloud
alice_wc.generate(alice_novel)
```

```python
# Plot the word cloud
fig = plt.figure()
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Refined Word Cloud without 'Said'

```python
# Remove stop words 'Said' because it doesn't add much insight, and then regenerate word cloud

stopwords.add('said') 

# Create the wordCloud and set parameters
alice_wc = WordCloud(
    background_color='white',
    max_words=2000, # Maximum of 2000 word to prevent Word cloud from getting less insightful
    stopwords=stopwords
)

# Create word cloud
alice_wc.generate(alice_novel)

# Plot the word cloud
fig = plt.figure()
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
