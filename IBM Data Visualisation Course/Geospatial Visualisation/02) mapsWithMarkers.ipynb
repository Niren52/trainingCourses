```python
# Import modules needed for the mapping
import numpy as np
import pandas as pd
import folium
from folium import plugins
```

```python
# Import data to a dataframe
df_incidents = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
# Can be downloaded from here: https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Police_Department_Incidents_-_Previous_Year__2016_.csv
```

```python
# Get Summary
df_incidents.head()
```

```python
# Get the first thousand points to plot for testing to see if working and reduce rendering time
limit = 1000
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
# create a map for the area of San Francisco
sanfran_map = folium.Map(location=[37.77, -122.42], zoom_start=11)
```

```python
#loop through dataframe and set parameters for the map markers
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category): # Get latitude, longitude and category of crime
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map) # At the end of each loop, add point to marker 


sanfran_map
```

# Alternative Option to segment different regions of San Francisco by number of crimes

```python
# New Instance of San Frnacisco Map
sanfran_map = folium.Map(location=[37.77, -122.42], zoom_start=11)

# Loop again but allow popup now
incidents = plugins.MarkerCluster().add_to(sanfran_map)
for lat, lng, label, in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

sanfran_map
```
