# Explanation of Map

### The map below has the latest borders, and borders are updated and the geolocation JSON is fitted to the new map.
### However, the Immigration data of Canada has some disparity between previous names and official names, therefore some countries display an all black colour, signifying there is no accurate data available. 
### Where possible, I have manually altered country names in the JSON geofile,  for countries of significance (i.e. the UK and Iran), since they have large migration figures. and are large in size, I also did the same for Greenland, to be part of the Kingdom of Denmark.

```python
world_geo = r'world-countries.json'
world_map = folium.Map(location=[0, 0], zoom_start=2, tiles='Mapbox Bright')
# Create choropleth map of Canada Immigration
world_map.choropleth(
    geo_data=world_geo,
    data=df_canada,
    columns=['OdName', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Immigration to Canada'
)

# Show Data
world_map
```
