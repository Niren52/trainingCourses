```
# Import Models needed for exercise
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Import the data file we will be training model against
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/kerasRefresher/kerasData/1.LinearRegression/IowaHousingPrices.csv')
```

```
# Get example rows of what summary of what the data looks like
df.head()
```

```
# Seperate values so that they can be fit to a model
squareFeet = df[['SquareFeet']].values
salesPrice = df[['SalePrice']].values
```

```
type(squareFeet) # Check type of data
```

```
model = keras.Sequential() # Create a sequential linear model
model.add(keras.layers.Dense(1, input_shape = (1,))) #Add relevant neural network layer
model.compile(keras.optimizers.Adam(lr = 1), 'mean_squared_error') # Use Adam Optimization Algorithm for gradient descent
model.fit(squareFeet, salesPrice, epochs = 30, batch_size = 10) # Train the model
```

```
#plot the graph showing the relationship
df. plot(kind ='scatter', x='SquareFeet', y='SalePrice', title = 'Linear Regression: Housing and Square Footage of Iowa Homes Relationship')
y_pred = model.predict(squareFeet)
plt.plot(squareFeet, y_pred, color = 'red')
```

```
# Model prediction for a Squarefootage of 5000
newSF = 5000
print(model.predict([newSF]))
```
