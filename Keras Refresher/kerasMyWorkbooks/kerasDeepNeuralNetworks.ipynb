```
# Import relevant modules
import keras
import numpy as np
import pandas as pd
```

```
# Load the data to a dataframe
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/kerasRefresher/kerasData/2.DNNs/HousingPrices.csv')
```

```
#Get example data
df.head()
```

```
#Drop the dependent variable so that we can have variables for the independent variables
x = df.drop(columns = ['SalePrice'])

# Check that this has been updated
x.head()
```

```
# Separate the dependent variable for use in modelling
Y = df[['SalePrice']]

#Check
Y.head()
```

```
model = keras.Sequential() # Create a linear Sequential model

#Create a network: 8-8-1 with Rectified Linear Unit
model.add(keras.layers.Dense(8, activation='relu', input_shape = (8,)))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss = 'mean_squared_error')
```

```
#Train model with callbacks and stop early if there isn't much change in fitting
model.fit(x,Y, epochs = 30, callbacks = [keras.callbacks.EarlyStopping(patience = 5)])
```

```
# Let's test with random data point to see what model fits towards
test_data = np.array([2003, 854, 1710, 2, 1, 3, 8, 2008])

#Make prediction with first column by reshaping the test input to a column:
print(model.predict(test_data.reshape(1,8), batch_size = 1))
```
